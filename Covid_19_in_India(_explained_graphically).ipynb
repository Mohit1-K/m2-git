```
import pandas as pd
import numpy as np
from google.colab import files
```

```
import matplotlib.pyplot as plt
% matplotlib inline
```

```
 url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
df1 = pd.read_csv(url)
df1.head()
```

```
Df=df1.loc[df1["iso_code"] =="IND"].copy()
```

```
Df
```

```
Confirmed_cases= Df["new_cases"].copy()
```

```
Deaths= Df["total_deaths"].copy()
```

```
Date= Df["date"].copy()
```

```
 
plt.figure(figsize=(10,7))
plt.plot(Date, Deaths, color= "red")
plt.ylabel("Total number of Deaths", fontweight= "bold")
plt.xlabel("(Months (2020) \n This graph shows the speed of increasing COVID-19 cases by months.)", fontweight= "bold")
 
 
positions = (1,35.36,65.72, 96.08, 126.44, 160.8, 194.16, 231.52, 266.88, 300.24)
labels = ("Jan", "Feb", "Mar", "Apt", "May", "Jun", "Jul", "Aug", "Sep", "Oct")
plt.xticks(positions, labels)
 
 
plt.show()
 
plt.savefig('plot.png', dpi=800, bbox_inches='tight')
files.download('plot.png')
```

```
 
 
plt.figure(figsize=(10,7))
plt.plot(Date, Confirmed_cases, color= "red")
plt.ylabel("Maximum number of cases per day", fontweight= "bold")
plt.xlabel("Months (2020)", fontweight= "bold")
positions = (1,35.36,65.72, 96.08, 126.44, 160.8, 194.16, 231.52, 266.88, 300.24)
labels = ("Jan", "Feb", "Mar", "Apt", "May", "Jun", "Jul", "Aug", "Sep", "Oct")
plt.xticks(positions, labels)
 
 
 
plt.show()
 
plt.savefig('plot_cases.jpg', dpi=800, bbox_inches='tight')
files.download('plot_cases.jpg')
```
